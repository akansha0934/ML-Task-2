```python
#importing 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
ds=pd.read_csv("Salary_Data.csv")
ds
```

```python
ds.columns
```

```python
#divide dataset a independant (input) and Dependant (output)
x = ds[['YearsExperience']]
y = ds[['Salary']]
```

```python
x
```

```python
y
```

```python
#plot graph
plt.scatter(x,y)
plt.show()
```

```python
#splitting dataset as training and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
# Introduce SLR Model

from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
#predict
result=reg.predict([[9]])
result
```

```python
result=reg.predict(x_test)
result
```

```python
y_test
```

```python
#visualize plot
plt.scatter(x,y)
result=reg.predict(x)
plt.plot(x, result, color='pink')
plt.show()
```

```python
# For User Input

years = int(input("Enter year of exp. : "))
myinput = pd.DataFrame(columns =['YearsExperience'], data=[[years]])
result = reg.predict(myinput)
print("Your Salary must be close to", round(result[0,0]))
```
